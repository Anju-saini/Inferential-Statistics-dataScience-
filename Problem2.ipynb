**Problem 2**

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain;

 Import Required Libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

%matplotlib inline

```

Define Distribution Parameters and X-axis Range

```python

average_strength = 5
variation_strength = 1.5

strength_range = np.arange(0, 10, 0.1)
distribution_curve = norm.pdf(strength_range, average_strength, variation_strength)
```

2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
threshold_1 = 3.17
portion_under_3_17 = norm.cdf(threshold_1, loc=average_strength, scale=variation_strength)
print(f" Bags weaker than {threshold_1} kg/cm²: {round(portion_under_3_17 * 100, 2)}%")

# Visualizing the area under the curve
plt.figure(figsize=(9,5))
plt.plot(strength_range, distribution_curve)
plt.fill_between(strength_range, 0, distribution_curve, where=(strength_range < threshold_1), color='red', alpha=0.4)
plt.axvline(threshold_1, color='red', linestyle='--')
plt.title("Bags with Strength < 3.17 kg/cm²")
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Density")
plt.grid(True)
plt.show()

```

**Bags weaker than 3.17 kg/cm²: 11.12%**

2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

```python
limit_2 = 3.6
portion_above_3_6 = 1 - norm.cdf(limit_2, loc=average_strength, scale=variation_strength)
print(f" Bags stronger than or equal to {limit_2} kg/cm²: {round(portion_above_3_6 * 100, 2)}%")

plt.figure(figsize=(9,5))
plt.plot(strength_range, distribution_curve)
plt.fill_between(strength_range, 0, distribution_curve, where=(strength_range >= limit_2), color='green', alpha=0.4)
plt.axvline(limit_2, color='green', linestyle='--')
plt.title("Bags with Strength ≥ 3.6 kg/cm²")
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

**Bags stronger than or equal to 3.6 kg/cm²: 82.47%**

2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

# Define Distribution Parameters and X-axis Range
average_strength = 5
variation_strength = 1.5
strength_range = np.arange(0, 10, 0.1)
distribution_curve = norm.pdf(strength_range, average_strength, variation_strength)

lower_limit = 5
upper_limit = 5.5

```

# Calculate the proportion between the limits

```python

portion_between_5_5_5 = norm.cdf(upper_limit, loc=average_strength, scale=variation_strength) - norm.cdf(lower_limit, loc=average_strength, scale=variation_strength)
print(f" Bags between {lower_limit} and {upper_limit} kg/cm²: {round(portion_between_5_5_5 * 100, 2)}%")

```

# Visualizing the area under the curve between the limits

```python

plt.figure(figsize=(9, 5))
plt.plot(strength_range, distribution_curve)
plt.fill_between(strength_range, 0, distribution_curve, where=((strength_range >= lower_limit) & (strength_range <= upper_limit)), color='orange', alpha=0.4)
plt.axvline(lower_limit, color='blue', linestyle='--', label=f'{lower_limit} kg/cm²')
plt.axvline(upper_limit, color='orange', linestyle='--', label=f'{upper_limit} kg/cm²')
plt.title(f"Bags with Strength Between {lower_limit} and {upper_limit} kg/cm²")
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

**Bags between 5 and 5.5 kg/cm²: 13.06%**

2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
min_range = 3
max_range = 7.5
outside_range = 1 - (norm.cdf(max_range, loc=average_strength, scale=variation_strength) - norm.cdf(min_range, loc=average_strength, scale=variation_strength))
print(f" Bags with strength outside 3 to 7.5 kg/cm²: {round(outside_range * 100, 2)}%")

plt.figure(figsize=(9,5))
plt.plot(strength_range, distribution_curve)
plt.fill_between(strength_range, 0, distribution_curve, where=(strength_range < min_range), color='purple', alpha=0.4)
plt.fill_between(strength_range, 0, distribution_curve, where=(strength_range > max_range), color='purple', alpha=0.4)
plt.axvline(min_range, color='purple', linestyle='--')
plt.axvline(max_range, color='purple', linestyle='--')
plt.title("Bags Outside the Strength Range 3–7.5 kg/cm²")
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Density")
plt.grid(True)
plt.show()

```

**Bags with strength outside 3 to 7.5 kg/cm²: 13.9%**
